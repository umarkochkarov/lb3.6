Дана функция $f(x) = x^2 \cdot \sin(x)$. Необходимо построить трехмерный график этой функции на интервале от -5 до 5.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x):
    return x**2 * np.sin(x)
```

```python
x = np.linspace(-5, 5, 100)  
y = f(x)                    
X, Y = np.meshgrid(x, y)    
Z = f(X)   
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('График функции f(x) = x^2 * sin(x)')
plt.show()
```
